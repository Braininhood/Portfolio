```python
import pandas as pd

# Step 1: Load the dataset directly from the URL
url = 'https://github.com/futureCodersSE/working-with-data/raw/main/Happiness-Data/2015.xlsx'
df = pd.read_excel(url)

# Step 2: Initial exploration
print("First few rows of the dataset:")
print(df.head())

print("\nData structure and missing values:")
df.info()

print("\nSummary statistics:")
print(df.describe())

print("\nCheck for missing values:")
print(df.isnull().sum())

print("\nColumn names:")
print(df.columns)

# Step 3: Clean the data (Renaming columns)
df.rename(columns={
    'Country': 'Country',
    'Region': 'Region',
    'Happiness Rank': 'Happiness_Rank',
    'Happiness Score': 'Happiness_Score',
    'Standard Error': 'Standard_Error',
    'Economy (GDP per Capita)': 'Economy_GDP_per_Capita',
    'Family': 'Family',
    'Health (Life Expectancy)': 'Health_Life_Expectancy',
    'Freedom': 'Freedom',
    'Trust (Government Corruption)': 'Trust_Gov_Corruption',
    'Generosity': 'Generosity',
    'Dystopia Residual': 'Dystopia_Residual'
}, inplace=True)

print("\nFirst few rows after renaming columns:")
print(df.head())

# Step 4: Formulate and answer the first question - Top 10 happiest countries
top_10_happiest = df[['Country', 'Happiness_Rank', 'Happiness_Score']].sort_values(by='Happiness_Rank').head(10)
print("\nTop 10 happiest countries:")
print(top_10_happiest)

# Step 5: Formulate and answer the second question - Correlation between GDP per capita and happiness score
correlation_gdp_happiness = df['Economy_GDP_per_Capita'].corr(df['Happiness_Score'])
print("\nCorrelation between GDP per capita and happiness score:")
print(correlation_gdp_happiness)

# Step 6: Formulate and answer the third question - Life expectancy and happiness across regions
region_life_expectancy_happiness = df.groupby('Region')[['Health_Life_Expectancy', 'Happiness_Score']].mean().sort_values(by='Happiness_Score', ascending=False)
print("\nAverage life expectancy and happiness score by region:")
print(region_life_expectancy_happiness)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset directly from the URL
url = 'https://github.com/futureCodersSE/working-with-data/raw/main/Happiness-Data/2015.xlsx'
df = pd.read_excel(url)

# Step 2: Initial exploration
print("First few rows of the dataset:")
print(df.head())

print("\nData structure and missing values:")
df.info()

print("\nSummary statistics:")
print(df.describe())

print("\nCheck for missing values:")
print(df.isnull().sum())

print("\nColumn names:")
print(df.columns)

# Step 3: Clean the data (Renaming columns)
df.rename(columns={
    'Country': 'Country',
    'Region': 'Region',
    'Happiness Rank': 'Happiness_Rank',
    'Happiness Score': 'Happiness_Score',
    'Standard Error': 'Standard_Error',
    'Economy (GDP per Capita)': 'Economy_GDP_per_Capita',
    'Family': 'Family',
    'Health (Life Expectancy)': 'Health_Life_Expectancy',
    'Freedom': 'Freedom',
    'Trust (Government Corruption)': 'Trust_Gov_Corruption',
    'Generosity': 'Generosity',
    'Dystopia Residual': 'Dystopia_Residual'
}, inplace=True)

print("\nFirst few rows after renaming columns:")
print(df.head())

# Step 4: Formulate and answer the first question - Top 10 happiest countries
top_10_happiest = df[['Country', 'Happiness_Rank', 'Happiness_Score']].sort_values(by='Happiness_Rank').head(10)
print("\nTop 10 happiest countries:")
print(top_10_happiest)

# Visualization: Bar plot of the top 10 happiest countries
plt.figure(figsize=(10, 6))
sns.barplot(x='Happiness_Score', y='Country', data=top_10_happiest, hue='Country', dodge=False, palette='viridis')
plt.title('Top 10 Happiest Countries (2015)')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()

# Step 5: Formulate and answer the second question - Correlation between GDP per capita and happiness score
correlation_gdp_happiness = df['Economy_GDP_per_Capita'].corr(df['Happiness_Score'])
print("\nCorrelation between GDP per capita and happiness score:")
print(correlation_gdp_happiness)

# Visualization: Scatter plot of GDP per Capita vs Happiness Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Economy_GDP_per_Capita', y='Happiness_Score', data=df, hue='Region', palette='Set2', s=100)
plt.title('GDP per Capita vs. Happiness Score (2015)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Step 6: Visualization: Heatmap of correlations between all numeric factors
numeric_df = df.select_dtypes(include='number')  # Select only numeric columns
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Happiness Factors')
plt.show()

# Step 7: Formulate and answer the third question - Life expectancy and happiness across regions
region_life_expectancy_happiness = df.groupby('Region')[['Health_Life_Expectancy', 'Happiness_Score']].mean().sort_values(by='Happiness_Score', ascending=False)
print("\nAverage life expectancy and happiness score by region:")
print(region_life_expectancy_happiness)

# Visualization: Bar plot of average life expectancy and happiness score by region
region_life_expectancy_happiness.plot(kind='bar', figsize=(12, 6), color=['skyblue', 'orange'])
plt.title('Average Life Expectancy and Happiness Score by Region (2015)')
plt.xlabel('Region')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend(title='Metrics')
plt.show()
```
