```python
# Exercise 1 - Line plot of net migration

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
data = pd.read_excel(url, sheet_name='Country Migration')

# Correct column names for net migration
net_migration_columns = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']

# Extract relevant columns for net migration
net_migration_df = data[net_migration_columns]

# Calculate the mean net migration for each year
mean_net_migration = net_migration_df.mean()

# Prepare data for Seaborn
plot_data = pd.DataFrame({
    'Year': mean_net_migration.index,
    'Mean Net Migration': mean_net_migration.values
})

# Set the style for the plot
sns.set_style("whitegrid")

# Create the line plot
plt.figure(figsize=(10, 6))
chart = sns.lineplot(data=plot_data, x='Year', y='Mean Net Migration', marker='o')

# Set x-axis labels and rotation
plt.xticks(ticks=plot_data['Year'], labels=plot_data['Year'], rotation=30)

# Add titles and labels
chart.set_title('Mean Net Migration Over the Years (2015-2019)')
chart.set_xlabel('Year')
chart.set_ylabel('Mean Net Migration')

# Show the plot
plt.show()

```

```python
# Exercise 2 - Creating a Bar chart for yearly migration

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
data = pd.read_excel(url, sheet_name='Country Migration')

# Correct column names for net migration
net_migration_columns = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']

# Extract relevant columns for net migration
net_migration_df = data[net_migration_columns]

# Calculate the maximum net migration for each year
max_net_migration = net_migration_df.max()

# Prepare data for Seaborn
plot_data = pd.DataFrame({
    'Year': ['2015', '2016', '2017', '2018', '2019'],
    'Maximum Net Migration': max_net_migration.values
})

# Set the style for the plot
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6))
chart = sns.barplot(x='Year', y='Maximum Net Migration', data=plot_data, palette='viridis')

# Add titles and labels
chart.set_title('Maximum Net Migration for the Years 2015-2019')
chart.set_xlabel('Year')
chart.set_ylabel('Maximum Net Migration')

# Show the plot
plt.show()
```

```python
# Exercise 3 - creating a bar graph using grouped data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
data = pd.read_excel(url, sheet_name='Country Migration')

# Extract the column for net migration in 2019
data_2019 = data[['target_country_wb_income', 'net_per_10K_2019']]

# Group by 'target_country_wb_income' and calculate the mean net migration for 2019
mean_migration_by_income = data_2019.groupby('target_country_wb_income').mean()

# Prepare data for Seaborn
plot_data = mean_migration_by_income.reset_index()
plot_data.columns = ['Income Level', 'Mean Net Migration']

# Set the style for the plot
sns.set_style("whitegrid")

# Create the horizontal bar plot
plt.figure(figsize=(12, 8))
chart = sns.barplot(x='Mean Net Migration', y='Income Level', data=plot_data, palette='viridis')

# Add titles and labels
chart.set_title('Mean Net Migration for 2019 by Income Level')
chart.set_xlabel('Mean Net Migration')
chart.set_ylabel('Income Level')

# Show the plot
plt.show()
```

```python
# Exercise 4 - Creating a pie chart of stress data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
psych_data = pd.read_csv(url)

# Group by 'firstlanguage' and calculate the mean stress level for each language
stress_data = psych_data.groupby('firstlanguage')['Stress'].mean()

# Extract labels and data
labels = stress_data.index
data = stress_data.values

# Create a color palette
colors = sns.color_palette('pastel')

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Mean Stress Levels by First Language')

# Display the plot
plt.show()

print("""
Commentary

The pie chart visualizes how stress levels are distributed among students who speak different languages. 
It provides insights into which language groups experience higher or lower stress levels on average. 
This can be useful for understanding the relationship between language and stress levels and may highlight areas where additional support or resources are needed.

In terms of insights:

    Proportions: The chart allows for quick comparison of stress levels across different language groups.
    High Stress: If one segment is notably larger, it indicates that students speaking that language experience higher average stress levels compared to others.
    Visual Clarity: The use of color palettes helps distinguish between different categories, making the chart more readable and aesthetically pleasing.

Using a pie chart for this kind of categorical data is effective as it visually represents the proportion of each category relative to the whole. 
If the number of categories is too large or the differences are subtle, a bar chart might be more informative.
""")
```

```python
# Exercise 7 - Create a scatterplot of Wellbeing and Stress with line of best fit

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the provided URL
url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
psych_data = pd.read_csv(url)

# Extract the relevant columns
x = psych_data['Stress']
y = psych_data['Wellbeing']

# Initialize the Seaborn style
sns.set(style="whitegrid")

# Create the scatter plot with a line of best fit
plt.figure(figsize=(10, 6))
chart = sns.regplot(x=x, y=y, scatter_kws={'s':50}, line_kws={'color':'red', 'label':'Fit Line'})

# Set axis labels
chart.set_xlabel('Stress')
chart.set_ylabel('Wellbeing')

# Add a title
plt.title('Scatterplot of Wellbeing vs. Stress with Line of Best Fit')

# Display the plot
plt.show()

print ("""
Data Story:
Description:

    The scatterplot illustrates the relationship between Wellbeing and Stress. Each point represents an individual's Wellbeing score and their corresponding Stress level.
    The red line represents the line of best fit, which is the result of a linear regression analysis. It shows the general trend or association between Wellbeing and Stress.

Observations:

    Trend: If the red line slopes upwards, it suggests that higher levels of Stress are associated with higher levels of Wellbeing. Conversely, a downward slope would suggest that higher 
    Stress is associated with lower Wellbeing.
    
    Cluster: You may notice clusters of points that deviate from the line. This can indicate variability in how individuals respond to stress and its impact on their Wellbeing.
    
    Fit: The line of best fit helps in understanding the general direction of the relationship. If the points are closely packed around the line, it indicates a stronger relationship between 
    Stress and Wellbeing. If the points are scattered widely, it suggests a weaker relationship.

Conclusion:

    The scatterplot with the line of best fit provides insights into the correlation between Stress and Wellbeing.
    A positive slope of the line would suggest that as Stress increases, so does Wellbeing, which might seem counterintuitive. In contrast, a negative slope suggests that higher Stress leads to lower Wellbeing, 
    which aligns with common expectations.
    Confidence in the conclusion can be strengthened by examining the scatterplot and assessing how well the data points align with the regression line. Additional statistical measures such as the correlation 
    coefficient or R-squared value could further quantify the strength of the relationship.

This approach combines visual and statistical analysis to understand the relationship between Stress and Wellbeing comprehensively.
""")
```

```python
# Student Exam Scores Data Analysis

print ("""
This notebook visualizes and analyzes the student exam scores dataset. The dataset contains information about the number of hours studied and the corresponding exam scores. 
We will explore the data through various types of visualizations.
""")
## Load Libraries and Data



import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Optional: Suppress warnings if they are not useful
warnings.filterwarnings('ignore', category=UserWarning, module='seaborn')

# Load the data
url = 'https://raw.githubusercontent.com/lilaceri/Working-with-data-/main/Data%20Sets%20for%20code%20divisio/student_scores.csv'
data = pd.read_csv(url)



## Data Description
print ("""
The dataset includes:
- `Hours`: The number of hours studied.
- `Scores`: The exam scores achieved by the students.

We will visualize the relationships between study hours and exam scores using various types of plots.
""")

# Line Plot of Exam Scores vs. Study Hours
plt.figure(figsize=(10, 6))
sns.lineplot(x='Hours', y='Scores', data=data, marker='o')
plt.title('Exam Scores vs. Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.tight_layout()  # Ensure everything fits into the figure area
plt.show()



## Line Plot of Exam Scores vs. Study Hours
print ("""
This line plot illustrates how exam scores vary with the number of study hours. Each point represents a student's score based on their study hours. 
The trend line helps visualize the overall relationship between study hours and exam scores. As study hours increase, exam scores tend to rise, indicating a positive correlation.
""")


# Bar Chart of Average Scores by Study Hours
average_scores = data.groupby('Hours')['Scores'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Hours', y='Scores', data=average_scores, palette='viridis')
plt.title('Average Exam Scores by Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Average Scores')
plt.tight_layout()  # Ensure everything fits into the figure area
plt.show()



## Bar Chart of Average Scores by Study Hours
print ("""
This bar chart displays the average exam scores for each study hour. It provides a clear view of how the average scores change as the study hours increase. 
Higher study hours generally lead to higher average scores, supporting the positive relationship observed in the line plot.
""")

# Scatter Plot of Scores vs. Study Hours with Regression Line
plt.figure(figsize=(10, 6))
sns.regplot(x='Hours', y='Scores', data=data, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title('Scatter Plot of Scores vs. Study Hours with Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.tight_layout()  # Ensure everything fits into the figure area
plt.show()



## Scatter Plot of Scores vs. Study Hours with Regression Line
print ("""
The scatter plot shows individual student scores against their study hours, with a regression line indicating the trend. 
The line of best fit confirms the positive correlation between study hours and exam scores. The scatter plot helps visualize the distribution and strength of this relationship.
""")


# Histogram of Exam Scores
plt.figure(figsize=(10, 6))
sns.histplot(data['Scores'], bins=10, color='skyblue', kde=True)
plt.title('Distribution of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.tight_layout()  # Ensure everything fits into the figure area
plt.show()



## Histogram of Exam Scores
print ("""
This histogram displays the distribution of exam scores among students. It shows how frequently different score ranges occur, with the Kernel Density Estimate (KDE) providing a smoothed curve 
to understand the distribution more clearly. Most scores are clustered around a central range, indicating the common score ranges among students.
""")


# Box Plot of Scores by Study Hours
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hours', y='Scores', data=data, palette='pastel')
plt.title('Box Plot of Exam Scores by Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.tight_layout()  # Ensure everything fits into the figure area
plt.show()



## Box Plot of Exam Scores by Study Hours
print ("""
The box plot provides a summary of the exam scores for each study hour category. It highlights the median, quartiles, and potential outliers. 
This visualization helps understand the spread and central tendency of scores across different study hours, indicating that study hours influence not only the average score but also the variability in scores.
""")

# Pair Plot of All Variables
sns.pairplot(data)
plt.show()



## Pair Plot of All Variables
print ("""
The pair plot shows the relationships between study hours and exam scores through scatter plots, histograms, and density plots. 
It provides insights into the bivariate relationships and distributions of both variables, making it easier to identify patterns and correlations.
""")
```
