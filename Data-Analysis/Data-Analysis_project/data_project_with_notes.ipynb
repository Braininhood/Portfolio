The function written, is_valid_credit_card_number(cc_num), checks if a given credit card number is valid based on its length and starting digits. Here’s a breakdown of how it works:
Steps Breakdown:

    Convert the Input to String:
        cc_num = str(cc_num) ensures the input is treated as a string, allowing for easy validation and manipulation (e.g., checking the length, first digits).

    Check if the Credit Card Number Consists of Only Digits:
        if not cc_num.isdigit(): This checks if the cc_num contains only numeric digits. If it contains non-numeric characters, the function returns False.

    Check the Length of the Credit Card Number:
        length = len(cc_num) calculates the length of the credit card number, which is important because different credit card providers have specific length requirements.
        American Express numbers are 15 digits long, while Visa, MasterCard, and Discover cards are typically 16 digits long.

    Validate Based on the Starting Digits:
        For American Express (Amex) cards, the function checks if the number starts with '34' or '37' (common Amex prefixes) and has a length of 15.
        For Visa, MasterCard, and Discover, it checks if the card has a length of 16 and starts with:
            '4' for Visa.
            '51', '52', '53', '54', or '55' for MasterCard.
            '6011' for Discover.

    Return False for Invalid Numbers:
        If the credit card number doesn’t match any of the valid patterns, the function returns False.

Why use this function:

    Basic Validation: This function provides a simple and quick way to validate credit card numbers by checking their length and starting digits.

    Use Case: It can be used in applications that handle credit card transactions to filter out clearly invalid numbers before further processing or validation steps (e.g., using the Luhn algorithm for deeper validation).

Example Usage:

print(is_valid_credit_card_number(4111111111111111))  # return True (valid Visa)
print(is_valid_credit_card_number(123456789012345))   # return False (invalid)

```python
# Define a function to check if a credit card number is valid
def is_valid_credit_card_number(cc_num):
    cc_num = str(cc_num)
    if not cc_num.isdigit():
        return False
    length = len(cc_num)
    if length == 15:  # American Express
        return cc_num.startswith(('34', '37'))
    elif length == 16:  # Visa, MasterCard, Discover
        return cc_num.startswith(('4', '51', '52', '53', '54', '55', '6011'))
    return False
```

The function get_payment_system(cc_num) determines the payment system (or card type) based on the credit card number. Here’s a breakdown of its functionality:
Steps Breakdown:

    Convert the Input to String:
        cc_num = str(cc_num): Ensures the credit card number is treated as a string for easier manipulation.

    Check Card Type Based on Prefix and Length:
        Visa:
            Checks if the number starts with '4' and has a length of 16 digits.
        MasterCard:
            Checks if the number starts with any of '51', '52', '53', '54', or '55' and has a length of 16 digits.
        American Express (Amex):
            Checks if the number starts with '34' or '37' and has a length of 15 digits.
        Discover:
            Checks if the number starts with '6011' and has a length of 16 digits.

    Return Payment System:
        Based on the checks, the function returns the corresponding payment system name ('Visa', 'MasterCard', 'American Express', or 'Discover').
        If none of the conditions match, it returns 'Incorrect card number', indicating that the card number does not fit any known patterns.

Why use this function:

    Card Identification: This function helps identify the type of credit card based on its number, which can be useful for categorizing transactions or routing them appropriately.

    Simplified Handling: Knowing the card type allows to apply specific business rules or processing logic based on the card issuer.

Example Usage:

This function covers common credit card types. If need to handle other card types or add more sophisticated checks (e.g., validating card numbers with the Luhn algorithm), can extend the function accordingly.

```python
# Define a function to determine the payment system based on the card number
def get_payment_system(cc_num):
    cc_num = str(cc_num)
    if cc_num.startswith('4') and len(cc_num) == 16:
        return 'Visa'
    elif cc_num.startswith(('51', '52', '53', '54', '55')) and len(cc_num) == 16:
        return 'MasterCard'
    elif cc_num.startswith(('34', '37')) and len(cc_num) == 15:
        return 'American Express'
    elif cc_num.startswith('6011') and len(cc_num) == 16:
        return 'Discover'
    else:
        return 'Incorrect card number'
```

The code provided performs the following tasks:
1. Apply the get_payment_system Function:

    Convert cc_num to String:
        df['cc_num'] = df['cc_num'].astype(str): Ensures the cc_num column is treated as a string to handle cases like leading zeros.
    Add payment_system Column:
        df['payment_system'] = df['cc_num'].apply(lambda x: get_payment_system(x)): Applies the get_payment_system function to each value in the cc_num column and creates a new column payment_system with the results. This column will contain the type of card or 'Incorrect card number'.

2. Sort the DataFrame:

    Define Custom Sorting Order:
        sort_order = { ... }: Specifies the desired order for sorting the payment systems. 'Visa' comes first, followed by 'MasterCard', 'American Express', 'Discover', and 'Incorrect card number' last.
    Add Temporary Sorting Column:
        df['sort_order'] = df['payment_system'].map(sort_order): Maps each payment system to its corresponding sort order value.
    Sort and Drop Temporary Column:
        df = df.sort_values(by='sort_order').drop(columns='sort_order'): Sorts the DataFrame by the temporary sort_order column and then removes the column.

Why use this code:

    Categorization: The payment_system column helps categorize each transaction by its card type, which can be useful for further analysis or reporting.
    Custom Sorting: Sorting the DataFrame based on a custom order ensures that the results are organized in a meaningful way. For instance, having 'Visa' transactions appear first might be useful for analysis or reporting purposes.

Result:

    After applying this code,  DataFrame will have a new column, payment_system, with card types or 'Incorrect card number'. The rows will be sorted so that Visa transactions come first, followed by MasterCard, American Express, Discover, and finally incorrect card numbers.

```python
# Apply the function to add the 'payment_system' column
if 'cc_num' in df.columns:
    # Convert 'cc_num' to string to avoid issues with leading zeros
    df['cc_num'] = df['cc_num'].astype(str)

    # Add the 'payment_system' column based on the card number
    df['payment_system'] = df['cc_num'].apply(lambda x: get_payment_system(x))

    print("\n'payment_system' column has been added with all rows retained, including incorrect card numbers.")

# Sort the DataFrame by 'payment_system' column
# Define custom sorting order: Visa first, Incorrect card number last
sort_order = {
    'Visa': 1,
    'MasterCard': 2,
    'American Express': 3,
    'Discover': 4,
    'Incorrect card number': 5
}

# Add a temporary column for sorting
df['sort_order'] = df['payment_system'].map(sort_order)

# Sort the DataFrame by the temporary column and then drop the temporary column
df = df.sort_values(by='sort_order').drop(columns='sort_order')
```
