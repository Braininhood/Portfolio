```python
import pandas as pd

# List of years to iterate over
years = list(range(1991, 2022))  # Include 2021

# Base URL for the datasets
base_url = "https://raw.githubusercontent.com/futureCodersSE/python-programming-for-data/main/Datasets/NBA-MVP/mvp_{}.csv"

# List to hold DataFrames
df_list = []

# Loop through the years and read each CSV into a DataFrame
for year in years:
    url = base_url.format(year)
    df = pd.read_csv(url)
    df['Year'] = year  # Add a column for the year
    df_list.append(df)

# Combine all the DataFrames into one
all_mvp_data = pd.concat(df_list, ignore_index=True)

# Display the first few rows of the combined DataFrame
print(all_mvp_data.head())

# Investigate the columns
print(all_mvp_data.columns)

# Sort by relevant metrics (e.g., share of MVP votes)
sorted_mvp_data = all_mvp_data.sort_values(by='Share', ascending=False)

# Display sorted data to check for trends
print(sorted_mvp_data.head(20))

# Group by players to see who has the most MVP votes over the years
player_trends = sorted_mvp_data.groupby('Player').agg({'Share': 'sum', 'Year': 'count'}).reset_index()
player_trends = player_trends.sort_values(by='Share', ascending=False)

# Display player trends
print(player_trends.head(10))
```
