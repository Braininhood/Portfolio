```python
# Exercise 1 - Histogram check

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true"
data = pd.read_csv(url)

# Check for missing values
print(data.info())
print(data.describe())

# Familiarize with the data
print(data.head())

# Plot histogram of 'Resilliance'
plt.figure(figsize=(10, 6))
plt.hist(data['Resilliance'], bins=10, edgecolor='black')
plt.title('Histogram of Resilience')
plt.xlabel('Resilience Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Exercise 2 - Probability Plot 

import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
url = "https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true"
data = pd.read_csv(url)

# Check for missing values and familiarize with the data
print(data.info())
print(data.describe())
print(data.head())

# Probability plot for 'Resilliance'
plt.figure(figsize=(10, 6))
stats.probplot(data['Resilliance'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Resilience')
plt.show()
```

```python
# Exercise 3 - Histogram of non-Gaussian data

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/presDirty.csv"
data = pd.read_csv(url)

# Check for missing values
print("Initial data info:")
print(data.info())

# Drop rows with missing values in the relevant column
data_cleaned = data.dropna(subset=['national_turnout'])

# Check the data again after dropping missing values
print("\nData info after dropping missing values:")
print(data_cleaned.info())

# Familiarize with the data
print("\nFirst few rows of the cleaned data:")
print(data_cleaned.head())

# Plot a histogram of 'national_turnout'
plt.figure(figsize=(10, 6))
plt.hist(data_cleaned['national_turnout'], bins=10, edgecolor='black')
plt.title('Histogram of National Turnout')
plt.xlabel('National Turnout')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Exercise 4 - Prob plot of non-Gaussian data

import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/Data%20Cleaning%20Data%20Sets/presDirty.csv"
data = pd.read_csv(url)

# Drop rows with missing values in the 'national_turnout' column
data_cleaned = data.dropna(subset=['national_turnout'])

# Generate and plot the probability plot
plt.figure(figsize=(10, 6))
stats.probplot(data_cleaned['national_turnout'], dist="norm", plot=plt)
plt.title('Q-Q Plot of National Turnout')
plt.show()
```

```python
# Exercise 5 - Linear regression on normal data

import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Load the dataset
url = "https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true"
data = pd.read_csv(url)

# Check for missing values and drop rows with missing values in relevant columns
data_cleaned = data.dropna(subset=['Resilliance', 'selfesteem'])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(data_cleaned['Resilliance'], data_cleaned['selfesteem'])

# Print the regression results
print(f"LinregressResult(slope={slope}, intercept={intercept}, rvalue={r_value}, pvalue={p_value}, stderr={std_err})")

# Plot scatter plot and line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned['Resilliance'], data_cleaned['selfesteem'], color='blue', label='Data points')
plt.plot(data_cleaned['Resilliance'], intercept + slope * data_cleaned['Resilliance'], color='red', label='Line of Best Fit')
plt.title('Scatter Plot with Line of Best Fit')
plt.xlabel('Resilience')
plt.ylabel('Self Esteem')
plt.legend()
plt.grid(True)
plt.show()
```
