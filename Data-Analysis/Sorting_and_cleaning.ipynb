```python
# Exercise 1 - get data, sort by happiness score

import pandas as pd

# Define the function to sort the data by Happiness Score in ascending order and return the first 5 rows
def sorted_rank(df):
    return df.sort_values(by="Happiness Score").head(5)

# Read the Excel file from the provided URL
url = "https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true"
happiness = pd.read_excel(url)

# Test the function
actual = sorted_rank(happiness).index[0]
expected = 157

# Check if the test passes
if actual == expected:
    print("Well done, test passed")
else:
    print("Test failed","Should have got", expected, "got", actual)

# Display the sorted dataframe
sorted_rank(happiness)
```

```python
# Exercise 2 - sort by multiple columns, display the first 5 rows

import pandas as pd

# Define the function to sort the data by Economy (GDP per Capita) and Health (Life Expectancy) in ascending order and return the first 5 rows
def get_gdp_health():
    url = "https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true"
    happiness = pd.read_excel(url)
    return happiness.sort_values(by=["Economy (GDP per Capita)", "Health (Life Expectancy)"]).head(5)

# Test the function
test = get_gdp_health()

actual = test['Happiness Score'].iloc[0]
expected = 4.517

# Check if the test passes
if actual == expected:
    print("Well done, test passed")
else:
    print("Test failed", "Should have got", expected, "got", actual)

# Display the sorted dataframe
get_gdp_health()
```

```python
# Exercise 3 - sorting in descending order

import pandas as pd

# Define the function to sort the data by Freedom and Trust (Government Corruption) in descending order and return the last 5 rows
def get_descending(df):
    sorted_df = df.sort_values(by=["Freedom", "Trust (Government Corruption)"], ascending=False)
    return sorted_df[["Country", "Region"]].tail(5)

# Read the Excel file from the provided URL
url = "https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true"
happiness = pd.read_excel(url)

# Test the function
actual = get_descending(happiness).index[0]
expected = 136

# Check if the test passes
if actual == expected and (len(get_descending(happiness)) == 5):
    print("Test passed", actual)
else:
    print("Test failed","Should have got", expected, "got", actual, "and length of series should have been 5 but was", len(get_descending(happiness)))

# Display the sorted dataframe
get_descending(happiness)
```

```python
# Exercise 4 - check for null values

import pandas as pd

# Define the function to check for NA values in the dataframe and print the result
def check_null(df):
    # Check if any NA values exist in the dataframe
    has_null = df.isnull().values.any()
    
    # Print the dataframe info to see which columns have null entries
    df.info()
    
    return has_null

# Read the CSV file from the provided URL
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv"
housing = pd.read_csv(url)

# Test the function
actual = check_null(housing)
expected = True

# Check if the test passes
if actual == expected:
    print("Test passed", actual)
else:
    print("Test failed", "Should have got", expected, "got", actual)

# Display the result of check_null function
print("Does the dataframe contain null values?", check_null(housing))

```

```python
# Exercise 5 - filter for non-null rows
import pandas as pd

# Define the function to filter non-null rows based on the number_of_jobs column
def filter_null(df):
    return df[df['number_of_jobs'].notna()]

# Read the CSV file from the provided URL
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv"
housing = pd.read_csv(url)

# Test the function
actual = len(filter_null(housing))
expected = 931

# Check if the test passes
if actual == expected:
    print("Test passed", actual)
else:
    print("Test failed expected", expected, "got", actual)

# Display the filtered dataframe
filtered_housing = filter_null(housing)
print(filtered_housing)

```

```python
# Exercise 5 - remove null values

import pandas as pd

# Define the function to remove rows with NA values in the life_satisfaction column
def drop_null(df):
    return df.dropna(subset=['life_satisfaction'])

# Read the CSV file from the provided URL
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv"
housing = pd.read_csv(url)

# Test the function
actual = drop_null(housing).index[0]
expected = 613

# Check if the test passes
if actual == expected and len(drop_null(housing)) == 352:
    print("Test passed", actual)
else:
    print("Test failed", "Should have got", expected, "got", actual, "and length of series should have been 352 but was", len(drop_null(housing)))

# Display the result of drop_null function
filtered_housing = drop_null(housing)
print(filtered_housing)
```

```python
# Exercise 6 - Removing duplicate entries

import pandas as pd

# Define the function to remove duplicate area entries keeping the first instance
def drop_duplicates(df):
    return df.drop_duplicates(subset=['area'], keep='first')

# Read the CSV file from the provided URL
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv"
housing = pd.read_csv(url)

# Test the function
actual = drop_duplicates(housing).index[0]
expected = 0

# Check if the test passes
if actual == expected and len(drop_duplicates(housing)) == 51:
    print("Test passed", actual)
else:
    print("Test failed","Should have got", expected, "got", actual, "and length of series should have been 51 but was", len(drop_duplicates(housing)))

# Display the result of drop_duplicates function
filtered_housing = drop_duplicates(housing)
print(filtered_housing)
```
