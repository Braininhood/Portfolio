```python
# Exercise 1 - is age closely related to income?

# Import necessary libraries
import pandas as pd
from scipy.stats import linregress

# Load the data from the provided URL
url = "https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/Income-Data.xlsx?raw=true"
county_level_df = pd.read_excel(url, sheet_name='county-level')

# Assign the 'Age' and 'Income' columns to variables x and y
x = county_level_df['Age']
y = county_level_df['Income']

# Perform linear regression using the linregress function
regression = linregress(x, y)

# Display the full regression result
print("Regression Result:", regression)

# Display individual components of the regression
print("Slope:", regression.slope)
print("Intercept:", regression.intercept)
print("R-value:", regression.rvalue)
print("P-value:", regression.pvalue)
print("Standard Error:", regression.stderr)
```

```python
# Exercise 3 - Does Stress predict Wellbeing?

# Import necessary libraries
import pandas as pd
from scipy.stats import linregress

# Load the dataset from the provided URL
url = "https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true"
df = pd.read_csv(url)

# Display the columns to understand the dataset structure
print(df.columns)

# Assign the 'Stress' and 'Wellbeing' columns to variables x and y
x = df['Stress']
y = df['Wellbeing']

# Perform linear regression using the linregress function
regression = linregress(x, y)

# Calculate r_squared from rvalue
r_squared = regression.rvalue ** 2

# Display the full regression result
print("Regression Result:", regression)

# Display individual components of the regression
print("Slope:", regression.slope)
print("Intercept:", regression.intercept)
print("R-value (correlation coefficient):", regression.rvalue)
print("R-squared:", r_squared)
print("P-value:", regression.pvalue)
print("Standard Error:", regression.stderr)

# Save slope and intercept in variables for later use
slope = regression.slope
intercept = regression.intercept

# Challenge: Write a function to check the significance of the regression
def check_significance(regression):
    p_value = regression.pvalue
    r_squared = regression.rvalue ** 2
    if p_value < 0.05:
        print(f"Model is significant, p < .05: {p_value}")
    else:
        print(f"Model is not significant, p >= .05: {p_value}")
    print(f"R-squared = {r_squared}")

# Test the function with the current regression result
check_significance(regression)
```

```python
# Create some charts and make some predictions from regressions

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Step 1: Load the dataset
url = "https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true"
df = pd.read_csv(url)

# Step 2: Perform linear regression for Stress vs Wellbeing
x_stress = df['Stress']
y_wellbeing = df['Wellbeing']
regression_stress_wellbeing = linregress(x_stress, y_wellbeing)

# Calculate the regression line
slope_stress_wellbeing = regression_stress_wellbeing.slope
intercept_stress_wellbeing = regression_stress_wellbeing.intercept
line_stress_wellbeing = slope_stress_wellbeing * x_stress + intercept_stress_wellbeing

# Step 3: Plot the regression result
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_stress, y=y_wellbeing, color='blue', label='Data Points')
plt.plot(x_stress, line_stress_wellbeing, color='red', label=f'Regression Line: y={slope_stress_wellbeing:.2f}x + {intercept_stress_wellbeing:.2f}')
plt.title('Regression of Wellbeing on Stress')
plt.xlabel('Stress')
plt.ylabel('Wellbeing')
plt.legend()
plt.show()

# Step 4: Make a prediction for Wellbeing given a specific Stress level
stress_value = 5  # Example stress level
predicted_wellbeing = slope_stress_wellbeing * stress_value + intercept_stress_wellbeing
print(f"Predicted Wellbeing for Stress level {stress_value}: {predicted_wellbeing:.2f}")

# Step 5: Plotting multiple regressions in subplots (example placeholders)
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Stress vs Wellbeing
sns.scatterplot(x=x_stress, y=y_wellbeing, ax=axs[0])
axs[0].plot(x_stress, line_stress_wellbeing, color='red')
axs[0].set_title('Stress vs Wellbeing')
axs[0].set_xlabel('Stress')
axs[0].set_ylabel('Wellbeing')

# Example: Age vs Income (replace with actual data if available)
# Assuming df has 'Age' and 'Income' columns
# Uncomment and modify this section if you have another dataset for Age vs Income

# x_age = df['Age']
# y_income = df['Income']
# regression_age_income = linregress(x_age, y_income)
# line_age_income = regression_age_income.slope * x_age + regression_age_income.intercept
# sns.scatterplot(x=x_age, y=y_income, ax=axs[1])
# axs[1].plot(x_age, line_age_income, color='red')
# axs[1].set_title('Age vs Income')
# axs[1].set_xlabel('Age')
# axs[1].set_ylabel('Income')

plt.tight_layout()
plt.show()

# Function to check significance of the regression model
def check_significance(regression):
    p_value = regression.pvalue
    r_squared = regression.rvalue ** 2
    if p_value < 0.05:
        print(f"Model is significant, p < .05: {p_value}")
    else:
        print(f"Model is not significant, p >= .05: {p_value}")
    print(f"R-squared = {r_squared}")

# Step 6: Test the significance of the Stress vs Wellbeing regression model
check_significance(regression_stress_wellbeing)
```
