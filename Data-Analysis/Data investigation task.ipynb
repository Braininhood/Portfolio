```python
# # Data investigation task

import pandas as pd

# Load the data from the URL into a DataFrame
url = 'https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv'
df = pd.read_csv(url)

# Display basic information about the DataFrame
print("Basic Information about the DataFrame:")
df_info = df.info()
print("\n")

# Display the first 5 rows of the DataFrame
print("First 5 rows of the DataFrame:")
df_head = df.head()
print(df_head)
print("\n")

# Convert 'mean_salary' to numeric, forcing errors to NaN
df['mean_salary'] = pd.to_numeric(df['mean_salary'], errors='coerce')

# Calculate the average of the 'mean_salary' and 'median_salary' columns for each area
print("Average 'mean_salary' and 'median_salary' by area:")
average_salaries_by_area = df.groupby('area').agg({
    'mean_salary': 'mean',
    'median_salary': 'mean'
}).reset_index()

print(average_salaries_by_area)
print("\n")

# Display summary statistics for the numerical columns
print("Summary Statistics for the numerical columns:")
df_describe = df.describe(include='all')
print(df_describe)
print("\n")

# Display the unique values in the 'area' column
print("Unique values in the 'area' column:")
unique_areas = df['area'].unique()
print(unique_areas)
print("\n")

# Display the last 5 rows of the DataFrame
print("Last 5 rows of the DataFrame:")
df_tail = df.tail()
print(df_tail)
print("\n")

# Select rows with indices from 10 to 15
print("Rows with indices from 10 to 15:")
df_selected_rows = df.iloc[10:16]
print(df_selected_rows)
print("\n")

# Display the column names
print("Column names in the DataFrame:")
column_names = df.columns
print(column_names)
```
