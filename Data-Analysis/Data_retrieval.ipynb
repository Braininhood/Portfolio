```python
# Exercise - read from an API
import pandas as pd
import requests

def get_api_data():
    url = "https://api.carbonintensity.org.uk/regional/scotland"
    json_data = requests.get(url).json()['data']
    return json_data

# Step 1-5: Fetch the data and inspect the first element
generation_data = get_api_data()
print(generation_data[0])  # Debug: Print the first element to inspect its structure

# Step 6: Normalize generation_data[0]
normalized_df = pd.json_normalize(generation_data[0])
display(normalized_df)  # Debug: Display the normalized DataFrame to inspect its structure

# Check if 'generationmix' is a column
if 'generationmix' in normalized_df.columns:
    # Step 7: Explode the 'generationmix' column
    exploded_df = normalized_df.explode('generationmix', ignore_index=True)

    # Display the resulting DataFrame
    display(exploded_df)

    # Extension: Normalize the 'generationmix' column further to separate its attributes
    generationmix_df = pd.json_normalize(exploded_df['generationmix'])

    # Combine the normalized 'generationmix' with the exploded DataFrame
    final_df = exploded_df.drop(columns=['generationmix']).join(generationmix_df)

    # Display the final DataFrame
    display(final_df)
else:
    print("The column 'generationmix' was not found in the normalized DataFrame.")
```
