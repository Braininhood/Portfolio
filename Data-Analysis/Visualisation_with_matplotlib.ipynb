```python
# Exercise 1 - Line plot of net migration
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
data = pd.read_excel(url, sheet_name='Country Migration')

# Correct column names for net migration
net_migration_columns = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']

# Create a new DataFrame with the correct net migration columns
net_migration_df = data[net_migration_columns]

# Calculate the mean net migration for each year
mean_net_migration = net_migration_df.mean()

# Prepare labels and values for plotting
years = mean_net_migration.index
mean_values = mean_net_migration.values

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(years, mean_values, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Mean Net Migration (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Mean Net Migration')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show the grid
plt.grid()

# Display the plot
plt.show()


```

```python
# Exercise 2 - Creating a Bar chart for yearly migration
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
data = pd.read_excel(url, sheet_name='Country Migration')

# Extract the relevant columns
net_migration_columns = ['net_per_10K_2015', 'net_per_10K_2016', 'net_per_10K_2017', 'net_per_10K_2018', 'net_per_10K_2019']
net_migration_df = data[net_migration_columns]

# Calculate the total net migration for each year
total_net_migration = net_migration_df.sum()

# Labels for the years
years = ['2015', '2016', '2017', '2018', '2019']
total_values = total_net_migration.values

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, total_values, color='skyblue')

# Add titles and labels
plt.title('Total Net Migration by Year (2015-2019)')
plt.xlabel('Year')
plt.ylabel('Total Net Migration')

# Display the bar chart
plt.show()
```

```python
# Exercise 3 - creating a bar graph using grouped data
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
url = 'https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true'
data = pd.read_excel(url, sheet_name='Country Migration')

# Group by income level and calculate the mean net migration for 2019
mean_migration_2019 = data.groupby('target_country_wb_income')['net_per_10K_2019'].mean()

# Extract labels (income levels) and values
income_levels = mean_migration_2019.index
mean_values = mean_migration_2019.values

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(income_levels, mean_values, color='skyblue')

# Add titles and labels
plt.title('Mean Net Migration in 2019 by Income Level')
plt.xlabel('Mean Net Migration')
plt.ylabel('Income Level')

# Show the plot
plt.show()
```

```python
# Exercise 4 - Creating a pie chart of stress data

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the provided URL
url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
psych_data = pd.read_csv(url)

# Group by 'firstlanguage' and calculate the mean stress level
mean_stress_by_language = psych_data.groupby('firstlanguage')['Stress'].mean()

# Extract data and labels
data = mean_stress_by_language
labels = data.index

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(labels))))

# Add a title
plt.title('Distribution of Stress Levels by Language')

# Display the plot
plt.show()
```

```python
# Exercise 7 - Create a scatterplot of Wellbeing and Stress with line of best fit

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load the data from the provided URL
url = 'https://github.com/lilaceri/Working-with-data-/blob/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv?raw=true'
psych_data = pd.read_csv(url)

# Extract the Stress and Wellbeing columns
x = psych_data['Stress']
y = psych_data['Wellbeing']

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data Points')

# Add x and y axis labels
plt.xlabel('Stress')
plt.ylabel('Wellbeing')
plt.title('Scatter Plot of Wellbeing vs. Stress')

# Perform linear regression
regression = linregress(x, y)
slope = regression.slope
intercept = regression.intercept

# Plot the line of best fit
plt.plot(x, slope*x + intercept, color='red', linewidth=2, label='Line of Best Fit')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Exercise 8 - Create a set of charts from a data set

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load the dataset
url = 'https://raw.githubusercontent.com/lilaceri/Working-with-data-/main/Data%20Sets%20for%20code%20divisio/student_scores.csv'
data = pd.read_csv(url)

# Bar Chart: Exam Scores Distribution
plt.figure(figsize=(10, 6))
plt.bar(data.index, data['Scores'], color='skyblue')
plt.xlabel('Index')
plt.ylabel('Scores')
plt.title('Distribution of Exam Scores by Index')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Histogram: Distribution of Scores
plt.figure(figsize=(10, 6))
plt.hist(data['Scores'], bins=10, color='green', edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Scores')
plt.grid(True)
plt.show()

# Box Plot: Statistical Summary of Scores
plt.figure(figsize=(10, 6))
plt.boxplot(data['Scores'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'))
plt.xlabel('Scores')
plt.title('Box Plot of Exam Scores')
plt.show()

# Scatter Plot: Scores vs. Hours Studied
plt.figure(figsize=(10, 6))
plt.scatter(data['Hours'], data['Scores'], color='blue', alpha=0.7)
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Scatter Plot of Scores vs. Hours Studied')
plt.show()

# Adding Line of Best Fit to Scatter Plot
regression = linregress(data['Hours'], data['Scores'])
slope = regression.slope
intercept = regression.intercept

plt.figure(figsize=(10, 6))
plt.scatter(data['Hours'], data['Scores'], color='blue', alpha=0.7, label='Data Points')
plt.plot(data['Hours'], slope * data['Hours'] + intercept, color='red', linewidth=2, label='Line of Best Fit')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Scatter Plot of Scores vs. Hours Studied with Line of Best Fit')
plt.legend()
plt.show()

```
