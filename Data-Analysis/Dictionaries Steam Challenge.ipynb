```python
# Task 4
# Try repeating task 3 but for different games, what about the last game or the 100th?


import pandas as pd

# Function to load data and get lists of users and titles
def get_users_and_titles():
    url = "https://drive.google.com/uc?id=1rkG8-cp-KLBc1zK4YMLHIsMMyyTVk5Ju"
    data_table = pd.read_csv(url)
    return data_table["User"].tolist(), data_table["Title"].tolist()

# Function to find users who bought a specific game title
def find_users_for_title(game_title, user_game_dict_list):
    user_list = []
    for user_game_dict in user_game_dict_list:
        for user, title in user_game_dict.items():
            if title == game_title:
                user_list.append(user)
    return user_list

# Get the lists of users and titles
users, titles = get_users_and_titles()

# Task 1: Create a list of dictionaries with user:title key-value pairs
user_game_dict_list = []
for user, title in zip(users, titles):
    user_game_dict = {user: title}
    user_game_dict_list.append(user_game_dict)

# Task 2: Create a list with only the unique titles
unique_titles = []
for title in titles:
    if title not in unique_titles:
        unique_titles.append(title)

# Task 4: Find users for different games
# For the first game
first_title = unique_titles[0]
first_title_users = find_users_for_title(first_title, user_game_dict_list)
print(f"First title: {first_title}")
print(f"Number of users who bought it: {len(first_title_users)}\n")

# For the 100th game
hundredth_title = unique_titles[99]  # 100th game (index 99)
hundredth_title_users = find_users_for_title(hundredth_title, user_game_dict_list)
print(f"100th title: {hundredth_title}")
print(f"Number of users who bought it: {len(hundredth_title_users)}\n")

# For the last game
last_title = unique_titles[-1]
last_title_users = find_users_for_title(last_title, user_game_dict_list)
print(f"Last title: {last_title}")
print(f"Number of users who bought it: {len(last_title_users)}\n")

# Optional: Print a few users from each list to inspect
print(f"First title users: {first_title_users[:10]}")
print(f"100th title users: {hundredth_title_users[:10]}")
print(f"Last title users: {last_title_users[:10]}")
```
