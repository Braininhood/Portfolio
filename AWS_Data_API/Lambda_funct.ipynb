```python
import json
from Func_calc import universal_math_operation

def is_digit(value):
    if isinstance(value, (int, float)):
        return True
    else:
        return False

def is_integer(value):
    try:
        int(value)
        return True
    except ValueError:
        return False

def is_positive_integer(value):
    try:
        num = int(value)
        return num > 0 and num == float(value)
    except ValueError:
        return False

def lambda_handler(event, context):
    if "body" in event.keys():
        request = event["body"]
        if type(request) is not dict:
            request = json.loads(request)
        
        if "data" in request.keys():
            data = request["data"]
            print("Received data:", data)  # Add this line to log the data

            # Check if num1 is provided
            if "num1" not in data:
                return {
                    'statusCode': 400,
                    'body': json.dumps({'error': 'num1 is required'})
                }

            # Check if num1 is strictly numeric
            if not is_digit(data["num1"]):
                return {
                    'statusCode': 400,
                    'body': json.dumps({'error': 'num1 must be strictly numeric'})
                }

            num1 = float(data["num1"])

            # Check if num1 is a positive integer for the factorial operation
            if "operation" in data and data["operation"] == "factorial" and not is_positive_integer(data["num1"]):
                return {
                    'statusCode': 400,
                    'body': json.dumps({'error': 'num1 for factorial must be a positive integer'})
                }

            # Initialize num2 before the conditional statements
            num2 = None

            # Check if num2 is provided for operations other than factorial
            if "operation" in data and data["operation"] == "factorial" and "num2" in data:
                return {
                    'statusCode': 400,
                    'body': json.dumps({'error': 'num2 is not required for the specified operation'})
                }



            # Check if num2 is provided for operations other than factorial
            if "operation" in data and data["operation"] != "factorial":
                if "num2" not in data:
                    return {
                        'statusCode': 400,
                        'body': json.dumps({'error': 'num2 is required for the specified operation'})
                    }

                    # Check if num2 is strictly numeric
                    if not is_digit(data["num2"]):
                        return {
                            'statusCode': 400,
                            'body': json.dumps({'error': 'num2 must be strictly numeric'})
                        }

                    num2 = float(data["num2"])
           
            operation = data.get("operation", "")

            # Valid operations
            valid_operations = ['add', 'subtract', 'multiply', 'divide', 'factorial']

            # Check if operation is in the valid operations list
            if operation not in valid_operations:
                return {
                    'statusCode': 400,
                    'body': json.dumps({'error': f'Invalid operation. Choose from: {valid_operations}'})
                }

            # Check if both num1 and num2 are provided for 'add', 'subtract', 'multiply', 'divide'
            if operation in ['add', 'subtract', 'multiply', 'divide']:
                if "num2" not in data:
                    return {
                     'statusCode': 400,
                     'body': json.dumps({'error': f'Both num1 and num2 are required for the {operation} operation'})
                             }

                # Check if num2 is strictly numeric
                if not is_digit(data["num2"]):
                    return {
                        'statusCode': 400,
                         'body': json.dumps({'error': 'num2 must be strictly numeric'})
                            }

                num2 = float(data["num2"])
            else:
                num2 = None


            # Perform the operation
            if operation == 'add':
                result = universal_math_operation('add', num1, num2)
            elif operation == 'subtract':
                result = universal_math_operation('subtract', num1, num2)
            elif operation == 'multiply':
                result = universal_math_operation('multiply', num1, num2)
            elif operation == 'divide':
                if num2 == 0:
                    return {
                        'statusCode': 400,
                        'body': json.dumps({'error': 'Cannot divide by zero'})
                    }
                result = universal_math_operation('divide', num1, num2)
            elif operation == 'factorial':
                result = universal_math_operation('factorial', int(num1), 0)  # num2 is not used for factorial

            # now the data has been collected, it can run the add_numbers(num1, num2) function setting return_data to the result, and set the statuscode to 200 (success)
            return_data = result
            statuscode = 200
        else:
            return_data = "Unable to get data"
            statuscode = 404
        # now it can return the result in a JSON object with some security settings in the headers
        return {
            'statusCode': statuscode,
            'headers': {
                'Content-Type': 'application/json',
                'Access-Control-Allow-Headers': 'Content-Type,X-Api-Key',
                'Access-Control-Allow-Methods': 'POST',
                'Access-Control-Allow-Origin': '*'
            },
            'body': json.dumps(return_data)
        }
```
